# Advertising Project using Logistic Regression 

In this project we will be working with artificial advertising data set, indicating whether or not a particular internet user clicked on an Advertisement. We will try to create a model that will predict whether or not they will click on an ad based off the features of that user.

```python
# This data set contains the following features:

#  'Daily Time Spent on Site': consumer time on site in minutes
#  'Age': cutomer age in years
#  'Area Income': Avg. Income of geographical area of consumer
#  'Daily Internet Usage': Avg. minutes a day consumer is on the internet
#  'Ad Topic Line': Headline of the advertisement
#  'City': City of consumer
#  'Male': Whether or not consumer was male
#  'Country': Country of consumer
#  'Timestamp': Time at which consumer clicked on Ad or closed window
#  'Clicked on Ad': 0 or 1 indicated clicking on Ad
```

# Logistic Regression

Now it's time to do a train test split, and train our model!


** Split the data into training set and testing set using train_test_split**

```python
from sklearn.model_selection import train_test_split
```

```python
ad_data.head()
```

```python
ad_data.columns
```

```python
X = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage','Male']]
```

```python
y = ad_data['Clicked on Ad']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33)
```

** Train and fit a logistic regression model on the training set.**

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
logmodel.coef_
```

## Predictions and Evaluations
** Lets predict values for the testing data.**

```python
predictions = logmodel.predict(X_test)
```

```python
predictions
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

** Creating a classification report for the model.**

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

**Conclusion**

** We can conclude the following from the above metrics **
1. Out of all the test data that our classifier predicted, 92% (141 out of 153) of the times it predicited correctly if the ad was clicked. Only 12 times the classifier got it wrong.
2. Out of all the test data that our classifier predicted, 93%(164 out of 177) of the times it predicited correctly if the ad was not clicked. Only 13 times the classifier got it wrong.
3. Out of all the test data in which ad clicked was 'yes', our classifer also predicted the same 92% of the times.
4. Out of all the test data in which ad clicked was 'no', our classifer also predicted the same 93% of the times.
